````python

        
        
def append_to_json(file_path, info):
    """Appends a dictionary to a JSON file, creating the file if it doesn't exist.

    Args:
        file_path (str): The path to the JSON file.
        info (dict): The dictionary containing the data to append.
    """

    try:
        if not os.path.exists(file_path):
            # File doesn't exist, create it with an empty list
            with open(file_path, 'w') as json_file:
                json.dump([], json_file) 

        with open(file_path, 'r+') as json_file:
            data = json.load(json_file)
            # Ensure the data is a list (to append items)
            if not isinstance(data, list):
                raise ValueError("JSON file must contain a list")
            
            data.append(info)
            json_file.seek(0)
            json.dump(data, json_file, indent=4)  # Format with indentation
            json_file.truncate()  # Remove any old data after the new content

    except json.JSONDecodeError:
        print(f"Invalid JSON format in: {file_path}")
    except ValueError as e:
        print(e) 
````

````python
import pandas as pd

def create_matrix_dataframe(ids1, ids2):
  """
  Creates a DataFrame with an 'id' column containing ids1, and other columns from ids2.
  Initially, all DataFrame elements (except 'id') are set to 0.

  Args:
      ids1: A list of unique IDs.
      ids2: Another list of unique IDs.

  Returns:
      A Pandas DataFrame with an 'id' column and columns corresponding to ids2.
  """

  # Create a DataFrame with ids2 as columns, filled with 0s
  df = pd.DataFrame(0, index=range(len(ids1)), columns=ids2)

  # Add the 'id' column with values from ids1
  df.insert(0, 'id', ids1) 

  return df



result_df = create_matrix_dataframe(first_id, second_in)

# Print the DataFrame
print(result_df)
````

````python
def compare_dataframes_with_labels(df1, df2):
  """
  Compares a "predict" DataFrame against a "label" DataFrame,
  highlighting differences with original and predicted values.

  Args:
      df1: The "predict" DataFrame.
      df2: The "label" DataFrame.

  Returns:
      A DataFrame detailing the differences.
  """

  # Ensure DataFrames have the same shape
  if df1.shape != df2.shape:
    raise ValueError("DataFrames must have the same shape for comparison.")


  # Calculate the difference
  difference_df = df1.ne(df2)

  # Get indices of differing elements
  diff_indices = difference_df.stack()[difference_df.stack()].index.tolist()

  # Create a DataFrame to store the results
  result_data = []
  for row, col in diff_indices:
    result_data.append({
        'Row': row,
        'Column': col,
        'Original Value (Label)': df2.loc[row, col],
        'Predicted Value (Predict)': df1.loc[row, col]
    })

  result_df = pd.DataFrame(result_data)

  return result_df

# Example usage (replace with your actual DataFrames)
````

````python
class matrix_evaluation:
    def __init__(self, label_df, predict_df):
        self.label = label_df
        self.predict_df = predict_df
    #this use for the matrix 293x81
    def extracting_data_verticle(self):
        result = {}
        df = self.predict_df.copy()
        # Iterate through columns (excluding 'id')
        for col in df.columns:
            if col != 'id':
                # Get indices where the value is 1
                indices_with_one = df[df[col] == 1].index
                # Append corresponding 'id' values to the result dictionary
            
                result[col] = df.loc[indices_with_one, 'id'].tolist()
                
        result2 = {}
        df2 = label.copy()
        # Iterate through columns (excluding 'id')
        for col in df2.columns:
            if col != 'id':
                # Get indices where the value is 1
                indices_with_one = df2[df2[col] == 1].index

                # Append corresponding 'id' values to the result dictionary
                result2[col] = df2.loc[indices_with_one, 'id'].tolist()
        
        return result, result2
    
    # this use for the matrix 81x293
    def extracting_data_horizontal(self):
        df = self.predict_df.copy()
        result= {}

        for index, row in df.iterrows():
            key = str(int(row['id']))
            result[key] = []
            for col_name in df.columns:
                #if col_name != 'id' and row[col_name] == 1:
                if col_name != 'id' and row[col_name] >= 0.5:
                    result[key].append(int(col_name))

        result2 = {}
        df2 = label.copy()
        # Iterate through columns (excluding 'id')
        for col in df2.columns:
            if col != 'id':
                # Get indices where the value is 1
                indices_with_one = df2[df2[col] == 1].index

                # Append corresponding 'id' values to the result dictionary
                result2[col] = df2.loc[indices_with_one, 'id'].tolist()
        
        return result, result2
    
    def check_relevant_retrieve(self, result, result2, type = 2):
        if type ==2:
            result, result2 = self.extracting_data_horizontal()
        else:
            result, result2 = self.extracting_data_verticle()
        accuracy_data = []

        for key in result2:  # Iterate through keys in the label dictionary (result2)
            label_values = set(result2[key])
            predict_values = set(result[key])  # Ensure we handle potential duplicates

            common_values = label_values.intersection(predict_values)
            accuracy = len(common_values) / len(label_values) if len(label_values) > 0 else 0

            accuracy_data.append({'Key': key, 'Accuracy': accuracy})

        accuracy_df = pd.DataFrame(accuracy_data)
        print(accuracy_df['Accuracy'].mean())
        return accuracy_df
    
    def check_precision(self, result, result2, type == 2):
       
        if type ==2:
            result, result2 = self.extracting_data_horizontal()
        else:
            result, result2 = self.extracting_data_verticle()
        accuracy_data = []
        for key in result2:
            label_values = set(result2[key])
            predict_values = set(result[key])
        
            # Calculate true positives (common values)
            true_positives = label_values.intersection(predict_values)

            # Calculate false positives (predicted but not in labels)
            false_positives = predict_values - label_values
            if (len(true_positives) + len(false_positives)) == 0 and len(label_values) != 0:
                accuracy = 0
            else:
                accuracy = len(true_positives) / (len(true_positives) + len(false_positives)) if len(label_values) > 0 else 0
            false_positive_rate = len(false_positives) / len(label_values) if len(label_values) > 0 else 0

            accuracy_data.append({
                'Key': key, 
                'Accuracy': accuracy, 
                'False Positive Rate': false_positive_rate
            })

        accuracy_df = pd.DataFrame(accuracy_data)
        print(accuracy_df['Accuracy'].mean())
        return accuracy_df
    
        
        
        
        
                
````
